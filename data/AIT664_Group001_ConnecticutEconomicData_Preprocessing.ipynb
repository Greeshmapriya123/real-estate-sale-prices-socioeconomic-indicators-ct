```python
import pandas as pd
df = pd.read_csv('/content/AIT664-Group001-Dataset05-Connecticut-Economic-Data-BusinessEstablishments.csv')
business_establishment_df = df.melt(id_vars=['Town'], var_name='Year', value_name='BusinessEstablishment')
business_establishment_df
```

```python
df = pd.read_csv('/content/AIT664-Group001-Dataset06-Connecticut-Economic-Data-EmploymentData.csv')
employee_df = df.melt(id_vars=['Town'], var_name='Year', value_name='EmploymentData')
employee_df
```

```python
df = pd.read_csv('/content/AIT664-Group001-Dataset07-Connecticut-Economic-Data-EmploymentWages.csv')
employee_wages_df = df.melt(id_vars=['Town'], var_name='Year', value_name='EmploymentWages')
employee_wages_df
```

```python
df = pd.read_csv('/content/AIT664-Group001-Dataset08-Connecticut-Economic-Data-UnemploymentRates.csv')
unemployment_df = df.melt(id_vars=['Town'], var_name='Year', value_name='UnemploymentRates')
unemployment_df.rename(columns={'Town/City': 'Town'}, inplace=True)
unemployment_df['Town'] = unemployment_df['Town'].str.title()
unemployment_df
```

```python
df = pd.merge(business_establishment_df, employee_df, on=['Town','Year'], how='left')
df = pd.merge(df, employee_wages_df, on=['Town','Year'], how='left')
df = pd.merge(df, unemployment_df, on=['Town','Year'], how='left')
```

```python
from google.colab import files
df.to_csv('AIT664-Group001-Dataset10-Connecticut-Economic-Data.csv', index=False)
files.download('AIT664-Group001-Dataset10-Connecticut-Economic-Data.csv')
```
